```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
#criação das 3 variáveis
#define as unidades de cada varíaveis 
#nb- ńivel da barragem, podendo variar de 0 at́é 20 metros
#nr- ńivel do rio, podendo variar de 0 at́é 12 metros
#nb- ńivel da barragem, podendo varia de 0 at́é 20 metros
#op - operação na barragem, considerando 8 comportas para a operação
x_nb = np.arange(0, 20, 1)
x_nr = np.arange(0, 12, 1)
x_op = np.arange(0, 10, 1)
```

```python
#escreve as variáveis
x_nb, x_nr, x_op
```

```python
#processo de fuzzificação da variável nb
y_nb_normal = fuzz.trimf(x_nb, [0, 0, 10])
y_nb_medio = fuzz.trimf(x_nb, [0, 10, 20 ])
y_nb_alto = fuzz.trimf(x_nb, [10, 20, 20])
```

```python
y_op_normal = fuzz.trimf(x_op, [0, 0, 4])
y_op_media = fuzz.trimf(x_op, [0, 4, 8])
y_op_alta = fuzz.trimf(x_op, [4, 8, 8])
```

```python
y_op_normal, y_op_media, y_op_alta
```

```python
fig, axP = plt.subplots(figsize=(8, 5))
axP.plot(x_op, y_op_normal, 'b', label='normal')
axP.plot(x_op, y_op_media, 'g', label='media')
axP.plot(x_op, y_op_alta, 'r', label='alta')
axP.set_title('Operação nas comportas')
axP.legend();
```

```python
nb_nivel_normal = fuzz.interp_membership(x_nb, y_nb_normal, 20,0)
nb_nivel_medio = fuzz.interp_membership(x_nb, y_nb_medio, 20,0)
nb_nivel_alto = fuzz.interp_membership(x_nb, y_nb_alto, 20,0)
```

```python
x_op0 = np.zeros_like(x_op)
x_op0
```

```python
fig, ax0 = plt.subplots(figsize=(16, 6))
ax0.fill_between(x_op, x_op0, ativa_op_normal, facecolor='b', alpha=0.7)
ax0.plot(x_op, y_op_normal, 'b', linewidth=0.5, linestyle='--',)
ax0.fill_between(x_op, x_op0, ativa_op_media, facecolor='g', alpha=0.7)
ax0.plot(x_op, y_op_media, 'g', linewidth=0.5, linestyle='--',)
ax0.fill_between(x_op, x_op0, ativa_op_alta, facecolor='r', alpha=0.7)
ax0.plot(x_op, y_op_alta, 'r', linewidth=0.5, linestyle='--',)
ax0.set_title('Saida das operações interligadas na barragem');
```

```python
controle = np.fmax(ativa_op_normal, np.fmax(ativa_op_media, ativa_op_alta))
```

```python
op = fuzz.defuzz(x_op,controle, 'centroid')
#comentários
op
```

```python
op_ativacao = fuzz.interp_membership(x_op,controle, op)
```

```python
op_ativacao
```

```python
fig, ax0 = plt.subplots(figsize=(16, 6))
ax0.plot(x_op, y_op_normal, 'b', linewidth=0.5, linestyle='--',)
ax0.plot(x_op, y_op_media, 'g', linewidth=0.5, linestyle='--',)
ax0.plot(x_op, y_op_alta, 'r', linewidth=0.5, linestyle='--',)
ax0.fill_between(x_op, x_op0, controle, facecolor='purple', linewidth=1.5, alpha=0.9)
ax0.plot([op, op], [0, op_ativacao], 'black', linewidth=0.5, alpha=0.9)
ax0.set_title('Resultado agregado após a defuzzificação (LINHA)');
print("(Operação na barragem (LINHA) - Resultado: ", op)
```
